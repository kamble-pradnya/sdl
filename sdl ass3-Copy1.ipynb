```python
import pandas as pd
import numpy as np

df=pd.read_csv("C:\datafile.csv")
df.head()

```

```python
import matplotlib.pyplot as plt

X=df["Year"]
Y=df["Count"]
plt.scatter(X,Y)
plt.title("Scatter plot")
plt.xlabel("Year")
plt.ylabel("count")
plt.show()
```

```python
import matplotlib.pyplot as plt
X=df["Month"]
Y=df["Amount"]
plt.bar(X,Y)
plt.title("Bar Plot")
plt.xlabel("Month")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt
X=df["Month"]
Y=df["Count"]
plt.plot(X,Y,marker="o",linestyle="-")
plt.title("Line Plot")
plt.xlabel("Month")
plt.ylabel("Count")
plt.show()
```
